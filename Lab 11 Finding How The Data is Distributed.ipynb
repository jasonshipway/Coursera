### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
job_sat_count = df['JobSat'].value_counts()

plt.figure(figsize=(10,10))
plt.pie(job_sat_count, labels=job_sat_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Job Satisfaction Distribution')
plt.axis('equal')
plt.show()
# Over 46% of respondents rate their job satisfaction between 8-10. 
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
have_worked = df['LanguageHaveWorkedWith'].dropna().str.split(',')
want_to_work = df['LanguageWantToWorkWith'].dropna().str.split(',')

from collections import Counter

have_counts = Counter([lang.strip() for sublist in have_worked for lang in sublist])
want_counts = Counter([lang.strip() for sublist in want_to_work for lang in sublist])

have_df = pd.DataFrame.from_dict(have_counts, orient='index', columns=['HaveWorkedWith'])
want_df = pd.DataFrame.from_dict(want_counts, orient='index', columns=['WantToWorkWith'])

lang_df = have_df.join(want_df, how='outer').fillna(0).astype(int)

lang_df_sorted = lang_df.sort_values(by='HaveWorkedWith', ascending=False).head(10)

lang_df_sorted.plot(kind='bar', figsize=(16, 8))
plt.title('Have Vs Used Languages')
plt.ylabel('Number of respondents')
plt.xticks(rotation=45)
plt.legend(title='Languages')
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
remote_region = pd.crosstab(df['Country'], df['RemoteWork'])
remote_country_pct = remote_region.div(remote_region.sum(axis=1), axis=0)

plt.figure(figsize=(12,8))
sns.heatmap(remote_region, fmt='d', cmap='YlGnBu')

plt.title('RemoteWork Distribution by Region')
plt.xlabel('Remote Work Type')
plt.ylabel('Region')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
print(df['JobSat'].head())
```

```python
## Write your code here
def clean_years_code(val):
    if val == 'Less than 1 year':
        return 0.5
    elif val == 'More than 20 years':
        return 21
    else:
        try:
            return float(val)
        except:
            return None
df['YearsCodeProClean'] = df['YearsCodePro'].apply(clean_years_code)

correlation_df = df[['JobSat', 'YearsCodeProClean']].dropna()

correlation = correlation_df.corr(method='pearson').loc['JobSat', 'YearsCodeProClean']
print(f"Pearson Correlation: {correlation:.3f}")
```
