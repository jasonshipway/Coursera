#### Step 1: Load the dataset


```python
file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"

df = pd.read_csv(file_path)
```


### Task 1: Exploring Relationships with Scatter Plots



#### 1. Scatter Plot for Age vs. Job Satisfaction



Visualize the relationship between respondents' age (`Age`) and job satisfaction (`JobSatPoints_6`). Use this plot to identify any patterns or trends.




```python
print(df['Age']).head()
```

```python
## Write your code here
df['Age'] = pd.to_numeric(df['Age'], errors='coerce')
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')
df = df.dropna(subset=['Age', 'JobSatPoints_6'])

plt.figure(figsize=(10, 6))
plt.scatter(df['Age'], df['JobSatPoints_6'], color='blue', edgecolor='black')
plt.title('Age vs Job Satisfaction')
plt.xlabel('Age')
plt.ylabel('Job Satisfaction')
plt.tight_layout()
plt.show()
```

#### 2. Scatter Plot for Compensation vs. Job Satisfaction


Explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`) using a scatter plot.


```python
## Write your code here
df['CompTotal'] = pd.to_numeric(df['CompTotal'], errors='coerce')
df = df.dropna(subset=['CompTotal'])

df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')
df = df.dropna(subset=['JobSatPoints_6'])

plt.figure(figsize=(10, 6))
plt.scatter(df['JobSatPoints_6'], df['CompTotal'], color='blue', edgecolor='black')
plt.title('Compensation vs Job Satisfaction')
plt.xlabel('Job Satifaction')
plt.ylabel('Compensation')
plt.tight_layout()
plt.show()
```
